```python
import pandas as pd

# --- 1. Data Preparation ---
# Load the dataset
try:
    df = pd.read_csv("trash_hauler_report.csv")
except FileNotFoundError:
    print("Error: 'trash_hauler_report.csv' not found. Please make sure the file is in the same directory.")
    # Create an empty DataFrame to avoid further errors
    df = pd.DataFrame(columns=['Request', 'Date Opened', 'Incident Address'])

# Clean column names by stripping whitespace
df.columns = df.columns.str.strip()

# Filter for rows that are actual missed pickups
missed = df[df["Request"].str.contains("Missed Pickup", case=False, na=False)].copy()

# Convert 'Date Opened' to datetime objects for calculations
missed["Date Opened"] = pd.to_datetime(missed["Date Opened"], errors='coerce')

# Clean up address strings
missed["Incident Address"] = missed["Incident Address"].str.strip().str.upper()

# Drop rows with invalid dates or addresses
missed.dropna(subset=["Date Opened", "Incident Address"], inplace=True)

# For our calculations, we only need unique dates for each address
unique_missed_dates = missed[["Incident Address", "Date Opened"]].drop_duplicates()


# --- 2. Alternative Method 1: Overlapping 180-day Windows ---

def calculate_fines_method_1(dates):
    """
    Calculates fines based on any three missed pickups within a 180-day period.
    Windows can overlap.
    """
    # Sort dates to create a timeline
    sorted_dates = sorted(dates)
    
    # Not enough missed pickups to qualify for a fine
    if len(sorted_dates) < 3:
        return 0
    
    fines = 0
    # Use a sliding window: check every consecutive set of 3 dates
    for i in range(len(sorted_dates) - 2):
        start_date = sorted_dates[i]
        end_date = sorted_dates[i+2] # The 3rd date in the window
        
        # If the 3rd pickup is within 180 days of the 1st, it's a fine
        if (end_date - start_date).days <= 180:
            fines += 1
            
    return fines

# Group by address and apply the function to the list of dates
fines_per_address_m1 = unique_missed_dates.groupby("Incident Address")["Date Opened"].apply(calculate_fines_method_1)

# Calculate the total fine amount
total_fines_method_1 = fines_per_address_m1.sum() * 1500


# --- 3. Alternative Method 2: Non-Overlapping 180-day Windows ---

def calculate_fines_method_2(dates):
    """
    Calculates fines based on three missed pickups within a 180-day period.
    Dates can only be used once to support a fine (non-overlapping).
    """
    sorted_dates = sorted(dates)
    
    if len(sorted_dates) < 3:
        return 0
        
    fines = 0
    i = 0
    while i < len(sorted_dates) - 2:
        start_date = sorted_dates[i]
        end_date = sorted_dates[i+2]
        
        if (end_date - start_date).days <= 180:
            # Fine found, increment count
            fines += 1
            # Jump past the 3 dates used for this fine
            i += 3
        else:
            # No fine found starting at this date, move to the next one
            i += 1
            
    return fines

# Group by address and apply the new function
fines_per_address_m2 = unique_missed_dates.groupby("Incident Address")["Date Opened"].apply(calculate_fines_method_2)

# Calculate the total fine amount
total_fines_method_2 = fines_per_address_m2.sum() * 1500


# --- 4. Comparison ---
# Get the original fine amount from your first notebook
original_total_fines = 762600 # From the result of the initial analysis

print("--- Comparison of Fine Calculation Methods ---")
print(f"Original Method Total: ${original_total_fines:,.0f}")
print(f"Alternative Method 1 Total: ${total_fines_method_1:,.0f}")
print(f"Alternative Method 2 Total: ${total_fines_method_2:,.0f}")
```
